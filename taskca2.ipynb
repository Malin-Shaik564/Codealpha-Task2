```python
# TASK 2: Unemployment Analysis with Python

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_excel(r"C:\Users\Shaik Farooq\OneDrive\Documents\Unemployment in India.csv.xlsx")

# Display first 5 rows
print("\nFirst 5 Rows:")
print(df.head())

# Display column names
print("\nColumn Names:")
print(df.columns)

# -----------------------------
# Data Cleaning
# -----------------------------

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop missing values
df.dropna(inplace=True)

# Convert date column
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Check duplicates
print("\nDuplicate rows:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

print("\nDataset cleaned successfully!")

# -----------------------------
# Data Exploration
# -----------------------------
print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# Average unemployment rate
avg_unemp = df['Estimated Unemployment Rate (%)'].mean()
print(f"\nAverage Unemployment Rate: {avg_unemp:.2f}%")

# -----------------------------
# Visualization 1:
# Unemployment trend over time
# -----------------------------
plt.figure(figsize=(12,6))
plt.plot(df['Date'],
         df['Estimated Unemployment Rate (%)'],
         marker='o')
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# -----------------------------
# Visualization 2:
# Covid Impact (Before vs During)
# -----------------------------
covid_date = '2020-03-01'

before_covid = df[df['Date'] < covid_date]['Estimated Unemployment Rate (%)'].mean()
during_covid = df[df['Date'] >= covid_date]['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(8,5))
plt.bar(['Before Covid', 'During Covid'],
        [before_covid, during_covid])
plt.title("Impact of Covid-19 on Unemployment")
plt.ylabel("Average Unemployment Rate (%)")
plt.show()

print(f"\nBefore Covid Avg: {before_covid:.2f}%")
print(f"During Covid Avg: {during_covid:.2f}%")

# -----------------------------
# Visualization 3:
# Region wise unemployment
# -----------------------------
region_avg = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(12,6))
region_avg.sort_values().plot(kind='bar')
plt.title("Average Unemployment by Region")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# -----------------------------
# Monthly Trend
# -----------------------------
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(10,5))
monthly_avg.plot(marker='o')
plt.title("Monthly Unemployment Trend")
plt.xlabel("Month")
plt.ylabel("Average Unemployment Rate (%)")
plt.grid()
plt.show()

print("\nAnalysis Completed Successfully!")
```
